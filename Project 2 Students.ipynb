#### d. Check for duplicate columns

There are a lot of columns in this data frame. Let's see if there are any are duplicates. Note that Pandas will not allow them to have the same exact column name, so they will likely be distinct on column name but will be copies otherwise.

```python
#
# Check for duplicate columns
#-----------
# I used Google to figure this out, and adapted this example for our purposes:  
# source: https://thispointer.com/how-to-find-drop-duplicate-columns-in-a-dataframe-python-pandas/ 
def getDuplicateColumns(df):
    '''
    Get a list of duplicate columns.
    It will iterate over all the columns in dataframe and find the columns whose contents are duplicate.
    :param df: Dataframe object
    :return: List of columns whose contents are duplicates.
    '''
    duplicateColumnNames = set()
    # Iterate over all the columns in dataframe
    for x in range(df.shape[1]):
        # Select column at xth index.
        col = df.iloc[:, x]
        # Iterate over all the columns in DataFrame from (x+1)th index till end
        for y in range(x + 1, df.shape[1]):
            # Select column at yth index.
            otherCol = df.iloc[:, y]
            # Check if two columns at x 7 y index are equal
            if col.equals(otherCol):
                duplicateColumnNames.add(df.columns.values[y])
    return list(duplicateColumnNames)

duplicateColumnNames = list(getDuplicateColumns(diabetes))
print('Duplicate Columns are as follows: ')
duplicateColumnNames
```

```python
#
# drop columns from duplicates list
#-----------
# now drop list of duplicate features from our df using the .drop() method
diabetes = diabetes.drop(columns=duplicateColumnNames) 
```

**MODEL DESCRIPTION(S):** 

**Linear Regression**: The linear regression model involves regressing the outcome on a set of predictors, and the model's success in fitting the data is determined by the residual sum of squares. The main assumptions of this model include normality of the residuals, linear relatinship between X and Y, and the errors are independently and identically distributed. If the true predictor response relationship is linear, the linear regression models are great models for producing estimate with low bias and low variance and will ultimately yield great results on the test dataset. However, if the relationship is non-linear, or if the sample size no longer greatly exceeds the number of predictors, this yields poor test performance. We can be confident to use this model for this analysis because of the linear relationships between diabetes and several features such as obesity and physical inactivity.  

**Ridge Regression**: The ridge regression model is an extension of the linear regression, which allows for some of the regression coefficients to shrink to zero using a tuning parameter called 'lambda.' This model provides an improvement from the linear regression model because it allows us to fit a model with all predictors instead of a subset as in linear regression, leading to a small increase in bias for a large variance decrease, and greater prediction accuracy. The disadvantage of this model is that we are fitting a model on all the predictors, and not all of them can be shrunk exactly to zero, which creates problems with interpretability. This method is particularly useful for this analysis due to the large amount of predictors in the dataset. 

**Lasso Regression**: The LASSO also shrinks regression coefficients to zero as in the ridge regression model, however, it allows the coefficients to be exactly zero, which offers an improvement from the ridge regression model by allowing for greater model interpretability. It also allows for variable selection unlike the ridge regression. One con would be the slight increase in bias over other methods.

```python
#
# RIDGE 
#-----------
# We will run 9*2*4 = 72 tests, each with 3-fold cross validation

# specify the hyperparameters
param_grid = {'alpha': np.arange(.1, 1, .1),
              'fit_intercept': [True, False],
              'solver': ['auto', 'svd', 'cholesky', 'lsqr']}

# execute the grid search
ridge_grid_reg = GridSearchCV(ridge_reg,  # model to be tuned
                              param_grid, # parameters to be searched as specified above
                              cv=3)       # 3-fold cross-validation to be used during hypertuning

# fit the tuning on the training data
ridge_grid_reg.fit(X_train, y_train)

# select the best performing model and predict with that on validation dataset 
best_index = np.argmax(ridge_grid_reg.cv_results_["mean_test_score"])  # find the best performing model
best_ridge_pred = ridge_grid_reg.best_estimator_.predict(X_test)   # find best estimator and predict on validate

# print the results  
print(ridge_grid_reg.cv_results_["params"][best_index])
print('Best CV R^2:', max(ridge_grid_reg.cv_results_["mean_test_score"]))
print('Validation R^2:', ridge_grid_reg.score(X_test, y_test))
print('Validation RMSE', rmse(best_ridge_pred, y_test))



```

1. What is bias-variance tradeoff? Why is it relevant to machine learning problems like this one?

Bias refers to an error in which the predicted outcome is systematically different from the true outcome. This problem typically arises from underfitting of the dataset. Variance tends to capture the noise of a model. If the variance is high, then the model estimates will change greatly across different training sets, which should not happen. High variance is typically the result of a model that overfits the data. 

The bias-variance tradeoff refers to the increase in bias as variance decreases and vice versa. A critical goal of machine learning is to identify the model which best deals with this tradeoff. For example, within the context of this project, we decided that the Ridge Regression performed well in predicting which counties were largely affected by Diabetes and thus should be the target for the diabetes prevention program. 2. Define overfitting, and why it matters for machine learning. How can we address it?

Overfitting occurs when there is a perfect fit of the model on the training data. This can lead to biased results because the model will most likely perform poorly on an independent test set. There are models with less flexible fitting approaches that avoid overfitting, including tree-based methods (bagging), forward stepwise selection, as well as Ridge and LASSO regression as used in this analysis.
